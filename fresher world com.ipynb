```python
link = "https://www.freshersworld.com/jobs/category/it-software-job-vacancies"
```

```python
jobs1 = uopen(link)
s = soup(jobs1.read())
```

```python
for container in containers:
    company_name = container.find("h3").text
    qualification = container.find("span",{"class":"qualifications display-block modal-open"}).text.replace(","," ")
    experience = container.find("span",{"class":"experience"}).text.text.replace(","," ")
    location = container.find("span",{"class":"job-location display-block modal-open"}).text.replace(","," ")
    post_date = container.find("div",{"class":"text-ago"}).text.replace(","," ")
    last_date = container.find("div",{"class":"col-md-4 col-xs-4 col-lg-4 padding-none"}).text.replace(","," ")
    
    print(company_name,qualification,experience,location,post_date,last_date,sep=" | ")
```

```python
jobs1 = open("jobs1.csv",'wt', encoding='utf-8')
jobs1.write('company_name|qualification|experience|location|post_date|last_date\n')

for container in containers:
    company_name = container.find("h3").text
    qualification = container.find("span",{"class":"qualifications display-block modal-open"}).text.replace(","," ")
    experience = container.find("span",{"class":"experience"}).text.replace(","," ")
    location = container.find("span",{"class":"job-location display-block modal-open"}).text.replace(","," ")
    post_date = container.find("div",{"class":"text-ago"}).text.replace(","," ")
    last_date = container.find("div",{"class":"col-md-4 col-xs-4 col-lg-4 padding-none"}).text.replace(","," ")
    
    print(company_name,qualification,location,post_date,last_date,sep=" | ")
    jobs1.write(f"{company_name}|{qualification}|{experience}|{location}|{post_date}|{last_date}\n")
jobs1.close()
```
